```python
sns.displot(data = new_penguins, x="flipper_length_mm", col="species", row="sex",
            binwidth=3, height=3, facet_kws=dict(margin_titles=True))
```

```python
#новый столбец, зависимый от другого - длина клюва high > 42, middle 35-42, short <35

new_penguins.loc[(new_penguins['bill_length_mm'] > 42), ['bill_length_category']] = 'long'
new_penguins.loc[((new_penguins['bill_length_mm'] <= 42) & (new_penguins['bill_length_mm'] >= 35)), ['bill_length_category']] = 'average'
new_penguins.loc[(new_penguins['bill_length_mm'] < 35), ['bill_length_category']] = 'short'
```

```python
sns.histplot(data = new_penguins, y = "bill_length_category", hue = 'species')
```

```python
new_penguins.rename({'bill_length_category': "bill_length_groups"}, axis = 1)
```

```python
sns.histplot(data = new_penguins, x = 'flipper_length_mm', hue = 'bill_length_groups')
```
