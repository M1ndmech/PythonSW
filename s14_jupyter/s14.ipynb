## Задание 3. 
Изучите взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров, используя сложенную или многорядовую столбчатую диаграмму
Процессоры от Samsung не изучайте

```python
cpu_table = pd.crosstab(df['Company'], df['Cpu_Company'])
cpu_df = cpu_table.drop(columns = ['Samsung']).reset_index()
cpu_df
```

### 3.1 Постройте график в абсолютных величинах

```python
plt.figure(figsize=(12,7))
n_ticks = np.arange(len(cpu_df['Company']))
offset = 0.2
w = 0.4
plt.bar(n_ticks - offset, cpu_df['AMD'], width = w)
plt.bar(n_ticks + offset, cpu_df['Intel'],width = w)
plt.xlabel('Company')
plt.ylabel('CPU count')
plt.xticks(n_ticks, cpu_df['Company'], rotation= 55)
plt.title('CPU count depending on brand')
plt.legend(['AMD','Intel']);
```

### 3.2 Постройте график в относительных величинах

```python
cpu_df_2 = pd.crosstab(df['Company'], df['Cpu_Company'], normalize="index").drop(columns = ['Samsung']).reset_index()

plt.figure(figsize=(10,5))

plt.bar(cpu_df_2['Company'], cpu_df_2['AMD'])
plt.bar(cpu_df_2['Company'], cpu_df_2['Intel'], bottom = cpu_df_2['AMD'])

plt.xlabel('Company')
plt.ylabel('CPU count')
plt.xticks(n_ticks, cpu_df_2['Company'], rotation= 55)
plt.title('CPU count normalized')
plt.legend(['AMD','Intel']);
```
